# Data visualization using matplotlib

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load data

df = pd.read_csv('my_data.csv')
```

```python
# Analytics task

TASK: conduct analytics on the marketing data to determine wwhich consumer demographics to target for specific products
    Build a dashboard to display insights
```

```python
## Explorator questions
1. What is the distribution of age groups
2. What is the income distribution
3. What is the average spend per product group by different age segments
4. What is the average spend per product group by different income segments
5. What is the average purchases by channel for different age groups
6..What is the average purchases by channel for different income groups
7. What is the total spend by age group
8. what is the total spend by income group
```

```python
# Configure Plotting canvas

plt.style.use('ggplot')

# creating matplotlib subplot structure to hold our plots

fig, axes = plt.subplots(4,3, figsize=(18,24))




#Q1: 1. what is the distribution of age groups?
# Cell1
age_group_counts = df['Age_Group'].value_counts().sort_index()
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, ax=axes[0,0] )

axes[0,0].set_title('Customers Age Distribution')
axes[0,0].set_ylabel('Customer Count')


#Cell2
#Q2:2. What is the income distribution?
income_group_counts = df['Income_Group'].value_counts().sort_index()
sns.barplot(x=income_group_counts.index, y=income_group_counts.values, ax=axes[0,1] )

axes[0,1].set_title('Customers Income Distribution')
axes[0,1].set_ylabel('Customer Count')





# Cell 3
#Q 3. What is the average spend per product group by different age segments

product_cats = ['MntWines', 'MntFruits','MntMeatProducts', 'MntFishProducts', 'MntSweetProducts','MntGoldProds']
avg_spend_Age = df.groupby('Age_Group')[product_cats].mean().transpose()
avg_spend_Age.plot(kind='bar', ax=axes[0,2], legend=False)
axes[0,2].set_title('Average Spending by Prodt Category(Age) ')
axes[0,2].set_ylabel('Average Spend on Prdts')
axes[0,2].legend(title='AgeGroup')
axes[0,2].tick_params(axis='x', rotation=45)


# Cell 4
#Q 4. What is the average spend per product group by different income segments

product_cats = ['MntWines', 'MntFruits','MntMeatProducts', 'MntFishProducts', 'MntSweetProducts','MntGoldProds']
avg_spend_Income = df.groupby('Income_Group')[product_cats].mean().transpose()
avg_spend_Income.plot(kind='bar', ax=axes[1,0], legend=False)
axes[1,0].set_title('Average Spending by Prodt Category(Income) ')
axes[1,0].set_ylabel('Average Spend on Prdts')
axes[1,0].legend(title='IncomeGroup')
axes[1,0].tick_params(axis='x', rotation=45)

# Cell5
#Q5What is the average purchases by channel for different age groups

channel_cats = ['NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases']
avg_purchase_Age = df.groupby('Age_Group')[channel_cats].mean().transpose()
avg_purchase_Age.plot(kind='bar', ax=axes[1,1], legend=False)
axes[1,1].set_title('Average Purchase by channel Category(Age) ')
axes[1,1].set_ylabel('Average Purchase on Chal')
axes[1,1].legend(title='AgeGroup')
axes[1,1].tick_params(axis='x', rotation=45)

#Q6What is the average purchases by channel for different income groups
channel_cats = ['NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases']
avg_purchase_Income = df.groupby('Income_Group')[channel_cats].mean().transpose()
avg_purchase_Income.plot(kind='bar', ax=axes[1,2], legend=False)
axes[1,2].set_title('Average Purchase by channel Category(Income) ')
axes[1,2].set_ylabel('Average Purchase on Chal')
axes[1,2].legend(title='IncomeGroup')
axes[1,2].tick_params(axis='x', rotation=45)


#Q7 What is the total spend by age group

product_cats = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Calculate the total spend for each individual by summing the specified product categories
df['Total_Spend'] = df[product_cats].sum(axis=1)

# Calculate the total spend by age group
total_spend_Age = df.groupby('Age_Group')['Total_Spend'].sum()
total_spend_Age.plot(kind='bar', ax=axes[2, 0], legend=False)
##sns.barplot(x='Age_Group', y='Total_Spend', data=total_spend_Age, ax=axes[2, 0], palette='tab20')
axes[2, 0].set_title('Total Spending by Age Group')
axes[2, 0].set_ylabel('Total Spend')
axes[2, 0].legend(title='AgeGroup')
axes[2, 0].tick_params(axis='x', rotation=45)


#Q8 What is the total spend by income group

product_cats = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Calculate the total spend for each individual by summing the specified product categories
df['Total_Spend'] = df[product_cats].sum(axis=1)

# Calculate the total spend by age group
total_spend_Income = df.groupby('Income_Group')['Total_Spend'].sum()
total_spend_Income.plot(kind='bar', ax=axes[2, 1], legend=False)
axes[2, 1].set_title('Total Spending by Income Group')
axes[2, 1].set_ylabel('Total Spend')
axes[2, 1].legend(title='IncomeGroup')
axes[2, 1].tick_params(axis='x', rotation=45)


#Q9What is the total spend per product by age group

product_cats = ['MntWines', 'MntFruits','MntMeatProducts', 'MntFishProducts', 'MntSweetProducts','MntGoldProds']
total_spend_Age = df.groupby('Age_Group')[product_cats].sum().transpose()
total_spend_Age.plot(kind='bar', ax=axes[2,2], legend=False)
axes[2,2].set_title('Total Spending by Prodt Category(Age) ')
axes[2,2].set_ylabel('Total Spend on Prdts')
axes[2,2].legend(title='AgeGroup')
axes[2,2].tick_params(axis='x', rotation=45)



#Q10 what is the total spend per product by income group
product_cats = ['MntWines', 'MntFruits','MntMeatProducts', 'MntFishProducts', 'MntSweetProducts','MntGoldProds']
total_spend_Income = df.groupby('Income_Group')[product_cats].sum().transpose()
total_spend_Income.plot(kind='bar', ax=axes[3,0], legend=False)
axes[3,0].set_title('Total Spending by Prodt Category(Income) ')
axes[3,0].set_ylabel('Total Spend on Prdts')
axes[3,0].legend(title='IncomeGroup')
axes[3,0].tick_params(axis='x', rotation=45)








```
